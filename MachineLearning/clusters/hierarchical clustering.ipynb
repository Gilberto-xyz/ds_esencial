```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv('C:/Users/G/OneDrive/Documentos/dataset/base_datos_2008.csv', nrows = 1e4)
```

```python
newdf = df[['AirTime', 'DepDelay']].dropna()
```

```python
newdf = df[['AirTime', 'Distance', 'TaxiOut', 'ArrDelay', 'DepDelay']].dropna()
```

```python
clstr = AgglomerativeClustering(n_clusters = 10)
clstr.fit(newdf)

# euclidean / L1/ L2/ manhattan/ cosine/ precomputed
# affinity = 'L1', linkage = 'complete'
```

```python
plt.scatter(newdf['AirTime'], newdf['DepDelay'], c = clstr.fit_predict(newdf))
plt.show()
```
