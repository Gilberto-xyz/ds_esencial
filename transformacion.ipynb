# Transformacion de la base de datos

```python
import pandas as pd
df = pd.read_csv('../dataset/base_datos_2008.csv')
df
```

```python
# Redondeamos los valores de ArrDelay en la nueva columna HoursDelay
df['HoursDelay'] = round( df['ArrDelay'] / 60)
```

```python
# imprimimos el valor de HoursDelay
df['HoursDelay']
```

```python
# Eliminamos la columna que acabamos de crear 
del( df['HoursDelay'])
```

```python
# Comprobamos si la columna se encuentra o no en el dataframe
df.head()
```

```python
# Eliminar columnas que no deseemos del dataframe
# Esto crea un dataframe nuevo de consulta, no actualiza el que ya teniamos
df.drop(['Diverted', 'Cancelled', 'Year'], axis=1)
```

```python
# Se puede eliminar tambien filas del dataframe
df.drop(0)
```

```python
# Podemos eliminar rangos de filas mediante drop
df.drop(range(1, 99))
```

```python
# Para poder guardar utilizaremos en la misma linea el inplace o igualaremos la variable donde esta nuestro dataframe 
df.drop(range(1, 99), inplace=True)
```

```python
df.head(10)
```

```python
# Guardar un dataframe solo con los origenes que hemos especificado
dfATL = df[ df['Origin'] == 'ATL']
dfHOU = df[ df['Origin'] == 'HOU']
```

```python
# Crear un dataframe que contiene los 2 dataframes anteriores
# Mediante append se pueden sumar ya que contienen la misma estructura de valores 
df_nuevo = dfATL.append(dfHOU)
df_nuevo.head(10)
```
