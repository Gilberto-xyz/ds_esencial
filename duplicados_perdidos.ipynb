# Tratar datos duplicados o perdidos

```python
# Libreria
import pandas as pd

# Dataset
# 10'6 = 1e6
df = pd.read_csv('../dataset/base_datos_2008.csv', nrows = 1e6)

# Dataframe
df
```

```python
# Dataframe artificial que duplica los datos
dfduplicado = df.append(df)
```

```python
# Sacamos la muestra de el dataframe para que no se trabaje con datos ordenados
dfduplicado = dfduplicado.sample(frac = 1)
```

### Eliminamos los duplicados y los guardamos en un nuevo dataframe 

```python

dflimpios = dfduplicado.drop_duplicates()
```

```python
# Mostramos el dataframe limpio (Sin elementos duplicados)
dflimpios
```

```python
# Comprobamos si la longitud del dataframe limpio es igual al dataframe con datos duplicados
len(dflimpios) == len(dfduplicado)
```

### Eliminar datos duplicados de una columna

```python
# Muestra solo 1 valor de la columna Dia del mes, en total solo 31 filas
dflimpios.drop_duplicates(subset = 'DayofMonth')
```

### Eliminar los datos faltantes del dataset

```python
df.dropna()
# Esto devuelve una consulta vacia por que hay vuelos donde al menos un caracter falta (NaN)
```

```python
# La mejor forma de llevar a cabo lo anterior y no ser tan radical es
df.dropna(thresh = 25)
# ahi considera 25 columnas de las 29 que se encuentran en total
```

```python
# Se puede complementar para que sea mas especifico y solo se considere como maximo los valores(n) faltantes
# Asi el dataframe va incrementando segun las nececidades que presentemos
df.dropna(thresh = len(df.columns)-1) 
# El resultado es el mismo por que las columnas van en bloque
```

### Filtrar por una columna en concreto

```python
# CancellationCode
# Vamos a filtrar mediante la columna donde no se tengan NaN 
df.dropna(subset = ['CancellationCode'])
```
