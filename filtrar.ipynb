# Filtrar datos en python 

```python
import pandas as pd

df = pd.read_csv('../dataset/base_datos_2008.csv')
df
```

```python
# Mostrar los primeros 20 elementos dentro de la columna ArrDelay
df['ArrDelay'].head(20)
```

```python
# Mostrar del registro 100 hasta el registro 200 
df[100 : 200]
```

```python
# Filtro avanzado
# Llamamos al dataframe con el que estamos trabajando y entre corchetes los valores que deseamos obtener 
df[ df['ArrDelay'] > 60].head()
```

```python
# Mostrar todos los vuelos que no se retrasaron 60 minutos
df[ df['ArrDelay'] != 60].head()
```

```python
# Mostrar todos los vuelos que tardaron igual o menos de 60 minutos
df[ df['ArrDelay'] <= 60]
```

```python
# Aplica con texto tambien

df[ df['Origin'] == 'ATL'].head()
```

```python
# Mostrar todos los vuelos de origen Houston o Atlanta
df[ df['Origin'].isin(['HOU', 'ATL'])].head()
```

```python
# Mostrar todos los registros que no tienen nada asignado en ArrDelay
# Dataset en el Dataframe
#   pandas is no asigned 
#            dataframe ArrDelay
df[ pd.isna( df['ArrDelay'])].head()
```
