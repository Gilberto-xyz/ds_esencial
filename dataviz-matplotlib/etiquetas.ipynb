# Etiquetas y leyendas en Matplotlib

```python
# Importamos las librerias necesarias
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos las primeras 100000 filas de los datos
df = pd.read_csv('C:/Users/G/Documents/dataset/base_datos_2008.csv', nrows = 100000)

# Tabla con el numero de vuelos que salieron cada dia de la semana
data =  np.unique(df.DayOfWeek, return_counts=True)
# Etiquetas (se usaran más adelante)
labs = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']

# Imprimimos los datos
data
```

## Grafico de sectores

```python
plt.pie( x= data[1],
        labels= data[0],
        radius= 1.5,
        colors= ['Red', 'Green', 'Orange', 'Blue', 'Gray', 'Pink', 'Black'],
        startangle= 90)

plt.show()
```

## Graficos personalizados

```python
plt.pie( x= data[1],
        labels= labs,
        radius= 1.5,
        # Seleccion de colores mediante una paleta
        colors= sns.color_palette('hls', 7),
        # Pay sobresalido
        explode= (0.1, 0.2, 0.2, 0.3, 0.1, 0.1, 0.1),
        # Valor en porcentaje
        autopct= '%1.1f%%',
        startangle= 90)

plt.show()
```

## Graficos con paquetes de seaborn

```python
plt = sns.barplot(x= labs, y= data[1])
plt.set(xlabel  = 'Día de la Semana', ylabel= 'Número de vuelos')
```
