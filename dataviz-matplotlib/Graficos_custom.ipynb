```python
# Librerias y forma de llamarlas en codigo
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

# Semilla para replicar los datos y obener los mismos valores
np.random.seed(0)
# BD de Aerol√≠neas
df = pd.read_csv("C:/Users/G/Documents/dataset/base_datos_2008.csv", nrows = 1000000)
# Remezclamos los datos y seleccionamos solo las primeras 100
df = df.sample( frac = 1).head(100)
```

```python
# Primer graf√≠co sin configuraci√≥n adicional 
plt.scatter( x = df.DayofMonth, y = df.ArrDelay, s = df.Distance)
plt.show()
```

```python
#            Dia del mes       Retraso de llegada      Distancia
plt.scatter( x = df.DayofMonth, y = df.ArrDelay, s = df.Distance,
#           Custom

#           Burbujas mas claras (0-1)
            alpha = 0.3,
#           Filtro para fines de semana (Morado = semana / Amarillo = fines de semana )
            c = df.DayOfWeek.isin([6, 7]))

# Titulo
plt.title('Retrasos en EEUU')

# Etiqueta de eje X
plt.xlabel('D√≠a del Mes')

# Etiqueta de eje Y
plt.ylabel('Retraso al Llegar')

# Limite de escala vertical (Para tiempos de llegada)
plt.ylim([0,135])

# Seleccionar el numero de marcas del mes
plt.xticks([0, 15, 30])

# texto para marcar un punto
plt.text(x=28, y=120, s='Mi vuelo')

plt.show()
```
